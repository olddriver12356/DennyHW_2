Suppose you have three different job offers with comparable salaries and job descriptions. ### 2.1 Decision stump by hand 
rubric={autograde}

**Your tasks:**

- With this toy dataset, build a decision stump (decision tree with only 1 split) manually, splitting on the condition `supportive_colleagues <= 0.5`. What training accuracy would you get with this decision stump? Save the accuracy as a decimal in an object named `supportive_colleagues_acc`. 

> You do not have to show any calculations or code. 

<div class="alert alert-warning">

Solution_2.1
    
</div>

_Points:_ 2

````python
df = train_df
left = df[df["supportive_colleagues"] <= 0.5] #s_c = 0
right = df [df["supportive_colleagues"] > 0.5]#s_c = 1

left_pred = left["target"].mode()[0]
right_pred = right["target"].mode()[0]

pred = df["supportive_colleagues"].apply(
    lambda v: left_pred if v <= 0.5 else right_pred
)

supportive_colleagues_acc = (pred == df["target"]).mean()

````
